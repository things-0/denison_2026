```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# redshift of the galaxy
r = 0.0582
red = 1 + r

#colours
cols = [
    'Wave', 'Open', 'gp1', 'rp1', 'ip1', 
    'zp1', 'yp1', 'wp1', 'Aero', 'Ray', 'Mol'
]

path = '' #put your path here

# Note: Ensure this path is correct for your machine
df = pd.read_csv(path, names=cols, delim_whitespace=True, comment='#', header=None, skiprows=27)

filter_colors = {
    'gp1': 'green',
    'rp1': 'red',
    'ip1': 'gold',
    'zp1': 'darkred',
    'yp1': 'black'
}


# 1. Initialize the figure FIRST
plt.figure(figsize=(10,6))

# 2. Plot the Filter Curves
for band, color in filter_colors.items():
    plt.plot(
        df['Wave'] * 10, 
        df[band], 
        label=band.replace('p1', ''),
        color=color, 
        lw=2, 
        alpha=0.8
    )

# 3. Plot the Vertical Lines (Rest Frame)
plt.axvline(x=6562.8, color='dodgerblue', linestyle='--', label=r'H$\alpha$ (Rest)')
plt.axvline(x=4861.35, color='orange', linestyle='--', label=r'H$\beta$ (Rest)')

# 4. Plot the Vertical Lines (Redshifted)
plt.axvline(x=6562.8 * red, color='red', linestyle=':', linewidth=2, label=r'H$\alpha$ (z=0.058)')
plt.axvline(x=4861.35 * red, color='pink', linestyle=':', linewidth=2, label=r'H$\beta$ (z=0.058)')

plt.title(f"PS1 Filter Response Curves vs Emission Lines (z={r})")
plt.xlabel(r"Wavelength ($\AA$)")
plt.ylabel("Response")
plt.grid(True, which='both', linestyle='--', alpha=0.5)

# Single legend for everything
plt.legend(title="Legend", loc='upper right')
plt.tight_layout()

plt.savefig('ps1.pdf')
plt.show()
```

```python
h_alpha_red_wavelength = 656.28 * red

g_transmission = np.interp(h_alpha_red_wavelength, df['Wave'], df['rp1'])
print(f"G: {g_transmission:.4f}")
```
