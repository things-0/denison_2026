ZTF Uses the PS1 calibration - this calibration ideally would be the exact same for ZTF

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

r = 0.0582
red = 1 + r

cols = [
    'Wave', 'Open', 'gp1', 'rp1', 'ip1', 
    'zp1', 'yp1', 'wp1', 'Aero', 'Ray', 'Mol'
]

df = pd.read_csv('/Users/shaunparasher/Documents/VS Code/denison_2026_test/shaun/Data/PS1 Calibration/apj425122t3_mrt.txt', names=cols, delim_whitespace=True, comment='#', header=None, skiprows=27)

plt.figure(figsize=(10, 6))

filter_colors = {
    'gp1': 'green',
    'rp1': 'red',
    'ip1': 'gold',
    'zp1': 'darkred',
    'yp1': 'black'
}

for band, color in filter_colors.items():
    plt.plot(
        df['Wave'], 
        df[band], 
        label=band.replace('p1', ''),
        color=color, 
        lw=2, 
        alpha=0.8
    )

plt.title("PS1 Filter Response Curves")
plt.xlabel(r"Lambda (nm)")
plt.ylabel(r"Xsection")
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.legend(title="Filter")
plt.tight_layout()
plt.axvline(x=656.28, color = 'dodgerblue', linestyle='--', label='H alpha')
plt.axvline(x=486.135, color = 'orange', linestyle='--', label='H beta')

plt.axvline(x=656.28 * red, color = 'red', linestyle='--', label='H alpha red')
plt.axvline(x=486.135 * red, color = 'pink', linestyle='--', label='H beta red')


plt.legend()
plt.show()
```

```python
h_alpha_red_wavelength = 656.28 * red

g_transmission = np.interp(h_alpha_red_wavelength, df['Wave'], df['rp1'])
print(f"G: {g_transmission:.4f}")
```
